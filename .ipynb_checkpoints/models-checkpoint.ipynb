```python
import pandas as pd
import seaborn as sns

# Read the CSV file into a DataFrame
hotels_poi_data = pd.read_csv(r'C:\Users\nas76\Desktop\Github Projects\Google Places\midtermProject\hotels_and_POIs.csv')

# Filter the DataFrame to include only 'hotels' and 'bedbreakfast' categories
filtered_data = hotels_poi_data[hotels_poi_data['hotel_categories'].isin(['hotels', 'bedbreakfast'])]

# Create the lmplot using the filtered data
sns.lmplot(x='hotel_rating', y='hotel_price', hue='hotel_categories', data=filtered_data)

```
