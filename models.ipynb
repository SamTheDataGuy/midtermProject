```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df_hotels_POIs = pd.read_csv(r'C:\Users\nas76\Desktop\Github Projects\Google Places\midtermProject\hotels_and_POIs.csv')

# Filter the DataFrame to include only 'hotels' and 'bedbreakfast' categories
filtered_data = df_hotels_POIs[df_hotels_POIs['hotel_categories'].isin(['hotels', 'bedbreakfast'])]

# Create the feature matrix X and target variable y
X = filtered_data[['hotel_rating']]
y = filtered_data['hotel_price']

# Add a constant term to the feature matrix X
X = sm.add_constant(X)

# Create a linear regression model
model = sm.OLS(y, X)

# Fit the model to the data
results = model.fit()

# Print the summary statistics
print(results.summary())

# Access specific statistics - P-values and R-squared
print("P-values:")
print(results.pvalues)
print("R-squared:", results.rsquared)

# Plot the actual data and the regression line
plt.scatter(X['hotel_rating'], y, color='blue', label='Actual')
plt.plot(X['hotel_rating'], results.fittedvalues, color='red', linewidth=2, label='Regression Line')
plt.xlabel('Hotel Rating')
plt.ylabel('Hotel Price')
plt.legend()
plt.show()

```
