```python
#Google Places API notebook

import pandas as pd
import requests
import os
```

```python
key = os.environ.get('api_key_places')
```

```python
#Getting information on Hotels in toronto using google places api

#url = f"https://maps.googleapis.com/maps/api/place/findplacefromtext/json?input=Museum%20of%20Contemporary%20Art%20Australia&inputtype=textquery&fields=formatted_address%2Cname%2Crating%2Copening_hours%2Cgeometry&key={api_key}"
url = f"https://maps.googleapis.com/maps/api/place/textsearch/json?query=hotels+in+toronto&key={key}"

response = requests.get(url)
response = response.json()
places_data = response['results']

df_google = pd.json_normalize(places_data)
df_google.head()
```

```python
df_google.to_csv('hotels_in_toronto.csv')
```
