```python
#Google Places API notebook

import pandas as pd
import requests
import os
```

```python
# Get API key
#key = os.environ.get('ta_key')

# For Sam, comment out below lines
from dotenv import load_dotenv
load_dotenv()
key = os.getenv('TRIPADVISOR_API_KEY')
```

```python
url = "https://tripadvisor16.p.rapidapi.com/api/v1/hotels/searchHotelsByLocation"

querystring = {
    "latitude": "43.68673", #I beleive we need to put {lat} for  "latitude"
    "longitude": "-79.59388", #{long}
    "checkIn": "2023-07-14", #required only future dates
    "checkOut": "2023-07-15", #required only future dates
    "pageNumber": "1",
    "currencyCode": "CAD",
    "radius":1000
}

headers = {
    "X-RapidAPI-Key": key,
    "X-RapidAPI-Host": "tripadvisor16.p.rapidapi.com"
}

response = requests.get(url, headers=headers, params=querystring)
#print(response.json())

data = response.json()

# Extract the list of hotels from the response data
hotels = data.get("data", {}).get("data", [])

selected_columns = ['id', 'title', 'primaryInfo', 'secondaryInfo', 'badge', 'isSponsored', 'provider', 'priceForDisplay', 'strikethroughPrice', 'priceDetails', 'priceSummary']

df = pd.DataFrame(hotels, columns=selected_columns)
df['badge'] = df['badge'].apply(lambda x: {'type': x['type'], 'year': x['year']} if isinstance(x, dict) and 'type' in x and 'year' in x else {})
# Print the DataFrame
df.head()
```
